# Einstein  Summation Conventions

### Einsum
This fumction is used to compute many multi-dimensional and linear alebric operations on arrays

```python
a = np.arange(1,10).reshape(3,3)
b = np.arange(21,30).reshape(3,3)
print(a)
print(b)
```

```python
#extract diagonals
np.einsum('ii->i',a)
```

```python
np.einsum('ii->i',b)
```

```python
#transpose
np.einsum('ij->ji',a)
```

```python
np.einsum('ij->ji',b)
```

```python
a.T
```

```python
b.T
```

```python
#determinant
np.linalg.det(a)
```

```python
np.linalg.inv(b)  #inverse
```

```python
#Matrix Multiplication
np.einsum('ij,jk',a,b)
```

# Matrix and Vector Multiplication

```python
A = np.array([[1,2], [3,4], [5,6]])
x = np.array([7,8])
print(np.einsum('ij,j->i', A, x))
```
